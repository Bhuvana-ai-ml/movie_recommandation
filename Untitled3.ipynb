```python
import pandas as pd
import numpy as np
import ast

from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.preprocessing import StandardScaler

from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    "movies_metadata.csv",
    engine="python",
    on_bad_lines="skip"
)

df = df[['title','genres','vote_average','popularity']]

df.head()
```

```python
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')

df.dropna(inplace=True)

df.reset_index(drop=True, inplace=True)

df.shape
```

```python
def extract_genres(x):
    try:
        genres = ast.literal_eval(x)
        return [i['name'] for i in genres]
    except:
        return []

df['genres'] = df['genres'].apply(extract_genres)

df[['title','genres']].head()
```

```python
mlb = MultiLabelBinarizer()

genre_encoded = mlb.fit_transform(df['genres'])

genre_df = pd.DataFrame(
    genre_encoded,
    columns=mlb.classes_
)

genre_df.head()
```

```python
features = pd.concat(
    [genre_df, df[['vote_average','popularity']]],
    axis=1
)

features.head()
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(features)
```

```python
inertia = []

for k in range(1,11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), inertia, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)

clusters = kmeans.fit_predict(X)

df['cluster'] = clusters

df[['title','cluster']].head()
```

```python
hc = AgglomerativeClustering(n_clusters=5)

df['hc_cluster'] = hc.fit_predict(X)
```

```python
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c=df['cluster'],
    cmap='viridis'
)

plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("Movie Clusters")

plt.show()
```

```python
tsne = TSNE(n_components=2, random_state=42)

X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(8,6))

plt.scatter(
    X_tsne[:,0],
    X_tsne[:,1],
    c=df['cluster'],
    cmap='viridis'
)

plt.title("t-SNE Movie Clusters")

plt.show()
```
